```python
## Problem 1: Chapter 3 - Exercise 5. Queue()
class Queue:
    
    def __init__(self):
        self.items = []
        
    def isEmpty(self):
        return self.items == []
    
    def enqueue(self, item):
        self.items.append(item)
    
    def dequeue(self):
        answer = self.items[0]
        self.items = self.items[1:]
        return answer
    
    def size(self):
        return len(self.items)
```

```python
## Problem 2: Chapter 3 - Exercise 19. UnorderedList class 
## with function Slice()
class Node:
    
    def __init__(self,initdata):
        self.data = initdata
        self.next = None

    def getData(self):
        return self.data

    def getNext(self):
        return self.next

    def setData(self,newdata):
        self.data = newdata

    def setNext(self,newnext):
        self.next = newnext
```

```python
class UnorderedList():
    
    def __init__(self):
        self.head = None
        
    def isEmpty(self):
        return self.head == None
    
    def add(self,item):
        temp = Node(item)
        temp.setNext(self.head)
        self.head = temp
        
    def size(self):
        current = self.head
        count = 0
        while current != None:
            count = count + 1
            current = current.getNext()
            
        return count
    
    def search(self,item):
        current = self.head
        found = False
        while current != None and not found:
            if current.data == item:
                found = True
            else:
                current = current.getNext()
        
        return found
    
    def remove(self,item):
        current = self.head
        previous = None
        found = False
        while not found:
            if current.getData() == item:
                found = True
            else:
                previous = current
                current = current.getNext()

        if previous == None:
            self.head = current.getNext()
        else:
            previous.setNext(current.getNext())
                
    def Slice(self,start,end):
        current = self.head
        now = 0
        answer = UnorderedList()
        while now != start:
            now = now + 1
            current = current.getNext()
        answer.add(current.getData())
        while now != (end-1):
            now = now + 1
            current = current.getNext()
            answer.add(current.getData())
        
        return answer
```

```python
## Problem 3: Chapter 3 - Programming Exercise 22. A stack using linked lists
##
class Stack:
    def __init__(self):
        self.data = UnorderedList()
        
    def isEmpty(self):
        return self.data.isEmpty()
        
    def push(self, item):
        self.data.add(item)
        
    def pop(self):
        answer = self.data.head.data
        self.data.remove(answer)
        return answer
    
    def peek(self):
        return self.data.head.data
    
    def size(self):
        return self.data.size()
```

```python
## Problem 4: Chapter 3 - Programming Exercise 23. A queue using linked lists
## Queue()
class Queue:
    
    def __init__(self):
        self.items = UnorderedList()
        
    def isEmpty(self):
        return self.items.isEmpty()
    
    def enqueue(self,item):
        self.items.add(item)
    
    def dequeue(self):
        previous = self.items.head
        size = self.items.size()
        if size == 0:
            answer = None
        elif size == 1:
            answer = previous.getData()
            self.items = UnorderedList()
        else:
            while size > 2:
                size = size-1
                previous = previous.getNext()
            last = previous.getNext()
            answer = last.getData()
            previous.setNext(None)
        
        return answer
        
    def size(self):
        return self.items.size()
```

```python
## Problem 5: Chapter 3 - Programming Exercise 24. A deque using linked lists
## Deque()
class Deque():
    
    def __init__(self):
        self.items = UnorderedList()
        
    def isEmpty(self):
        return self.items.isEmpty()
        
    def addFront(self, item):
        add = Node(item)
        current = self.items.head
        if current == None:
            self.items.add(item)
        else:
            while current.getNext() != None:
                current = current.getNext()
            current.setNext(add)
    
    def addRear(self, item):
        self.items.add(item)
        
    def removeFront(self):
        previous = self.items.head
        size = self.items.size()
        if size == 0:
            answer = None
        elif size == 1:
            answer = previous.getData()
            self.items = UnorderedList()
        else:
            while size > 2:
                size = size-1
                previous = previous.getNext()
            last = previous.getNext()
            answer = last.getData()
            previous.setNext(None)
        
        return answer
    
    def removeRear(self):
        current = self.items.head
        answer = current.getData()
        self.items.head = current.getNext()
        
        return answer
    
    def size(self):
        return self.items.size()
```

```python
## Problem 6: Chapter 3 - Adapted from Exercise 27. A new deque using new linked lists
# New Node with double references
class Node2():
    
    def __init__(self,initdata):
        self.data = initdata
        self.pre = None
        self.next = None

    def getData(self):
        return self.data
    
    def getPre(self):
        return self.pre

    def getNext(self):
        return self.next

    def setData(self,newdata):
        self.data = newdata
        
    def setPre(self,newpre):
        self.pre = newpre

    def setNext(self,newnext):
        self.next = newnext
```
