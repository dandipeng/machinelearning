```python
import pandas as pd
import numpy as np
```

```python
envirn = pd.read_excel('Environmental_characteristics_aquifers.xlsx',header=0)
agency = pd.read_excel('Agencies_before_SGMA.xlsx',header=0)
gsa = pd.read_excel('GSA.xlsx',header=0)
agency = agency.assign(gsaID=np.nan)
#gsa_bsn = pd.read_excel('GSAs_to_sub_basin.xlsx',header=0)
```

```python
ag_name_lower = list(map(lambda x: x.lower(), agency.agencyname))
```

```python
def corres(ckl, ttl, name):
    output = []
    for i in ckl:
        try: 
            rsl = ttl.index(i)
            agency.loc[rsl,'gsaID'] = name
            output.append((name, i))
        except: continue
    
    return output
```

```python
fnl = []
for i in range(gsa.iloc[:,53:77].shape[0]):
    names = list(set(map(lambda x: str(x).lower(), \
                         gsa.iloc[i,53:77][~gsa.iloc[:,53:77].isnull().iloc[i,:]])))
    corrp = corres(names,ag_name_lower, gsa.GSAID[i])
    fnl += corrp
    print(len(corrp))
```
